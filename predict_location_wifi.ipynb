```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
```

```python
data = pd.read_csv('train.csv', header=None)
data.columns = ['Id', 'AP1', 'AP2', 'AP3', 'AP4', 'AP5', 'AP6', 'AP7', 'Room']
data
```

```python
data.plot(kind="scatter", x="AP1", y="Room", label="AP 1");
data.plot(kind="scatter", x="AP2", y="Room", label="AP 2");
data.plot(kind="scatter", x="AP3", y="Room", label="AP 3");
data.plot(kind="scatter", x="AP4", y="Room", label="AP 4");
data.plot(kind="scatter", x="AP5", y="Room", label="AP 6");
data.plot(kind="scatter", x="AP7", y="Room", label="AP 7");
```

```python
sample_train = data.sample(frac=0.70)
sample_train
```

```python
sample_rest = data.loc[data.index.difference(sample_train.index)]
sample_rest
```

```python
X = np.array(sample_train.iloc[:,1:8].values)
X
```

```python
y = np.array(sample_train.iloc[:,-1].values)
y
```

```python
# from sklearn.svm import SVC
# clf = SVC()

from sklearn.naive_bayes import GaussianNB
clf = GaussianNB()

# from sklearn.neighbors import KNeighborsClassifier
# clf = KNeighborsClassifier(n_neighbors=5)

# from sklearn import linear_model
# clf = linear_model.LinearRegression()
```

```python
clf.fit(X,y)
```

```python
test_X = np.array(sample_rest.iloc[:,1:8].values)
test_y = np.array(sample_rest.iloc[:,-1].values)
```

```python
test_pred = clf.predict(test_X)
```

```python
from sklearn.metrics import accuracy_score

accuracy_score(test_y, test_pred)
```

```python
features_test = np.array(pd.read_csv('test.csv', header=None).iloc[:,1:8].values)
features_test
```

```python
y_pred = clf.predict(features_test)
y_pred
```

```python
id_df = pd.read_csv('test.csv', header=None).iloc[:,0:1]
id_df.columns = ['Id']
id_df
```

```python
pred_df = pd.DataFrame(y_pred)
pred_df.columns = ['Room']
pred_df
```

```python
ans_df = id_df
ans_df['Room'] = pred_df['Room'].values
ans_df
```

```python
ans_df.to_csv('submission.csv', index=False)
```
